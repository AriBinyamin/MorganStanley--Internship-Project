```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
%matplotlib inline
import yfinance as yf
import datetime
import pandas_datareader as pdr
get_ipython().run_line_magic('matplotlib', 'inline')
```

```python
start = datetime.datetime(2017,1,1)
end   = datetime.datetime(2022,7,25)
```

```python
Nasdaq = wb.DataReader("^IXIC",'yahoo',start,end)
```

```python
TechSector = wb.DataReader("^NDXT",'yahoo',start,end)
```

```python
Nasdaq['Open'].plot(label='Nasdaq',figsize=(15,7))
TechSector['Open'].plot(label='Technology Sector',figsize=(15,7))
plt.xlabel("Year")
plt.ylabel("Index Value")
plt.legend()
```

```python
Bio_Tech = wb.DataReader("^NBI",'yahoo',start,end)
Semi_conduct = wb.DataReader("^SOX", 'yahoo',start,end)
Tele_comms = wb.DataReader("^IXTC", 'yahoo',start,end)
Computer = wb.DataReader("^IXCO", 'yahoo',start,end)
```

```python
#Bio-Tech and nasdaq moving average 
Bio_Tech['MA_10'] = Bio_Tech.Close.rolling(10).mean()
Bio_Tech['MA_50'] = Bio_Tech.Close.rolling(50).mean()
Nasdaq['MA_10'] = Nasdaq.Close.rolling(10).mean()
Nasdaq['MA_50'] = Nasdaq.Close.rolling(50).mean()
```

```python
#Semi-conducter moving average
Semi_conduct['MA_10'] = Semi_conduct.Close.rolling(10).mean()
Semi_conduct['MA_50'] = Semi_conduct.Close.rolling(50).mean()
```

```python
#Tele-communications moving average 
Tele_comms['MA_10'] = Tele_comms.Close.rolling(10).mean()
Tele_comms['MA_50'] = Tele_comms.Close.rolling(50).mean()
```

```python
#Computer moving average 
Computer['MA_10'] = Computer.Close.rolling(10).mean()
Computer['MA_50'] = Computer.Close.rolling(50).mean()
```

```python
#50 day moving average for nasdaq and tech 
plt.figure(figsize=(15,10))
plt.grid(True)
plt.plot(Bio_Tech['MA_50'], label= 'BioTech MA 50 day')
plt.plot(Nasdaq['MA_50'], label='Nasdaq MA 50 day')
plt.plot(Semi_conduct['MA_50'], label='Semi-conductor MA 50 day')
plt.plot(Computer['MA_50'], label='Computer MA 50 day')
plt.plot(Tele_comms['MA_50'], label='Tele-comms MA 50 day')
plt.legend(loc=2)
```

```python
#10 day moving average for nasdaq and tech sector 
plt.figure(figsize=(15,10))
plt.grid(True)
plt.plot(Bio_Tech['MA_10'], label= 'BioTech MA 10 day')
plt.plot(Nasdaq['MA_10'], label='Nasdaq MA 50 day')
plt.plot(Semi_conduct['MA_10'], label='Semi-conductor MA 10 day')
plt.plot(Computer['MA_10'], label='Computer MA 10 day')
plt.plot(Tele_comms['MA_10'], label='Tele-comms MA 10 day')
plt.legend(loc=2)
```
